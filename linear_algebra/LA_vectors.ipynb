```python
def gram_schmidt_orthonormalization(V: np.ndarray):
    '''
    Constructs orthonormal basis orom a basis of space W 
    Inputs (shape (n,n)): V, where each column is a basis vector
    '''
    n = V.shape[0]
    V_T = V.T

    E = np.zeros((n,n))
    E[0] = V_T[0] / np.sqrt(np.sum(V_T[0]**2))
    R = np.zeros((n,n))
    for i in range(1,n):
        R[i] = V_T[i] - np.sum([np.dot(V_T[i], E[j]) * E[j] for j in range(i)], axis=0)
        #normalize residual 
        E[i] = R[i] / np.linalg.norm(R[i])
    
    return E.T
```

```python
def gram_schmidt_orthonormalization(V: np.ndarray) -> np.ndarray:
    """Construct an orthonormal basis from a set of linearly independent vectors.

    Parameters
    ----------
    V : np.ndarray, shape (n, n)
        Matrix whose columns are the input basis vectors.

    Returns
    -------
    E : np.ndarray, shape (n, n)
        Matrix whose columns form an orthonormal basis for the column space of V.

    Raises
    ------
    ValueError
        If a zero residual is encountered, indicating linearly dependent input vectors.
    """
    n = V.shape[1]
    E = np.zeros_like(V, dtype=float)

    for i in range(n):
        u = V[:, i].astype(float)
        for j in range(i):
            u -= np.dot(u, E[:, j]) * E[:, j]
        norm = np.linalg.norm(u)
        if norm < 1e-10:
            raise ValueError(f"Column {i} is linearly dependent on previous columns.")
        E[:, i] = u / norm

    return E
```
