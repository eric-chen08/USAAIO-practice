```python
import numpy as np
```

```python
def kernel_function(x):
    t = np.outer(x,x)
    return (1 + t + 2*(t**2))**2
```

```python
x = np.arange(5)
kernel_function(x)
```

# Copies And Views

* Simple assignments do not copy Numpy array objects
* If the new array is modified, the old array is modified also
* The view method creates a new array object that looks at the same data
* Slicing an array returns a view of it
* The copy method makes a complete copy of the array and its data
* The flatten() method makes a copy of a 2D array and flattens it as a 1D array
* The ravel() method creates a flattened view of the 2D array



```python
u = np.arange(1, 4)
v = u
print(v)
```

```python
u[0] = 10
print('u =', u)
print('v =', v)
```

```python
u = np.arange(1, 4)
w = np.copy(u)
u[0] = 10
print('u =', u)
print('w =', w)
```
