## Problem 5.5

```python
import numpy as np
```

```python
def kernel_function(x):
    phi_0 = np.ones(x.shape)
    phi_1 = np.sqrt(2) * x 
    phi_2 = np.sqrt(5) * (x ** 2)
    phi_3 = 2 * (x ** 3)
    phi_4 = 2 * (x ** 4)
    phi = np.vstack((phi_0, phi_1, phi_2, phi_3, phi_4))
    print("phi:")
    print(phi)

    K = np.matmul(phi.T, phi)
    return K
```

```python
x = np.array([1, 2, 3, 4, 5])
K = kernel_function(x)
print("K:")
print(K)
```

```python
def alternative_kernel(x):
    t = np.outer(x,x)
    return (1 + t + 2*(t**2))**2
```

```python
x = np.array([1, 2, 3, 4, 5])
K_alt = alternative_kernel(x)
print(K_alt)
```

# Problem 7 

```python
def compute_hyperplane(arr):
    #arr's shape is (2,d)
    x_1 = arr[0]
    x_2 = arr[1]

    midpoint = (x_1 + x_2)/2
    difference = (x_1 - x_2)
    theta = difference / (np.linalg.norm(difference))
    b = -(theta.T @ midpoint)
    return theta, b
```

```python
ages = df["age"].to_numpy()
y_yes = (df["y"] == "yes").to_numpy().astype(int)

# 10 bins => 11 edges
edges = np.linspace(ages.min(), ages.max(), 11)

ratios = []
bin_centers = []

for i in range(10):
    left, right = edges[i], edges[i+1]
    
    # include right edge only for the last bin to avoid dropping max-age rows
    if i < 9:
        mask = (ages >= left) & (ages < right)
    else:
        mask = (ages >= left) & (ages <= right)
    
    total = mask.sum()
    ratio = (y_yes[mask].sum() / total) if total > 0 else np.nan
    
    ratios.append(ratio)
    bin_centers.append((left + right) / 2)

plt.figure()
plt.plot(bin_centers, ratios, marker="o")
plt.xlabel("age bin (center)")
plt.ylabel("subscription ratio (y == 'yes')")
plt.title("Subscription ratio vs. age bin")
plt.show()
```
